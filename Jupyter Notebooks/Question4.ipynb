## Olympic data Analysis:

```python
#Olympic games medal dataset:
df_o2
```

```python
#Only 3 asian countries have hosted olympic games till now. So, we can analyse for those countries 
df_o2[df_o2['Country']==df_o2['Host_Country']]['Country'].value_counts()
```

```python
#Splitting data into 2 dataframes:
host=df_o2[df_o2['Country']==df_o2['Host_Country']].reset_index(drop=True) #Contains countries performance when they are host
non_host=df_o2[df_o2['Country']!=df_o2['Host_Country']].reset_index(drop=True) #Contains countries performance when they are not host
```

```python
host
```

```python
non_host
```

```python
#Overall analysis for all years combined:
host=host.groupby(['Year','Country'])['Gold','Silver','Bronze','Total'].agg(sum).reset_index()
```

```python
host
```

```python
host=host.groupby(['Country'])['Gold','Silver','Bronze','Total'].mean().reset_index()
#If a country has hosted for 2 years. Eg. Japan. Then, we'll aggreate it's medal tally by calculating their mean.
#Means if a country wins 6 golds in 2 years as host, then upon aggregating, that country wins 3 gold medals/year.
```

```python
host=host.rename(columns={'Gold':'Gold/Yr','Silver':'Silver/Yr','Bronze':'Bronze/Yr','Total':'Total/Yr'})
```

```python
host
```

```python
#Repeating same for non host countries:
non_host.groupby(['Year','Country'])['Gold','Silver','Bronze','Total'].agg(sum).reset_index()
```

```python
#Overall analysis:
non_host=non_host.groupby(['Year','Country'])['Gold','Silver','Bronze','Total'].agg(sum).reset_index()
```

```python
non_host
```

```python
non_host=non_host.groupby(['Country'])['Gold','Silver','Bronze','Total'].mean().reset_index()
```

```python
non_host
```

```python
non_host=non_host.rename(columns={'Gold':'Gold/Yr','Silver':'Silver/Yr','Bronze':'Bronze/Yr','Total':'Total/Yr'})
```

```python
non_host=non_host.sort_values(["Gold/Yr","Silver/Yr","Bronze/Yr"],ascending=False).reset_index(drop=True)
```

```python
non_host
```

```python
#Filtering data for only those Asian countries that have hosted Olympic games(China, Japan and South Korea):
non_host=non_host[:3]
```

```python
non_host
```

```python
host
```

**Now, we'll plot graphs on the basis of the data obtained by us:**

```python
#Total Medals won on an average as hosts and as non hosts comparison graphically:
plt.title("Total medals won as host country and as non host country per year:")
plt.bar(host['Country'],host['Total/Yr'],width=0.1,label='host')
plt.bar(non_host['Country'],non_host['Total/Yr'],width=0.1,label='non-host')
plt.legend()
plt.show()
```

```python
#Gold Medals won on an average as hosts and as non hosts comparison graphically:
plt.title("Gold medals won as host country and as non host country per year:")
plt.bar(host['Country'],host['Gold/Yr'],width=0.1,label='host')
plt.bar(non_host['Country'],non_host['Gold/Yr'],width=0.1,label='non-host')
plt.legend()
plt.show()
```

```python
#Silver Medals won on an average as hosts and as non hosts comparison graphically:
plt.title("Silver medals won as host country and as non host country per year:")
plt.bar(host['Country'],host['Silver/Yr'],width=0.1,label='host')
plt.bar(non_host['Country'],non_host['Silver/Yr'],width=0.1,label='non-host')
plt.legend()
plt.show()
```

```python
#Bronze Medals won on an average as hosts and as non hosts comparison graphically:
plt.title("Bronze medals won as host country and as non host country per year:")
plt.bar(host['Country'],host['Bronze/Yr'],width=0.1,label='host')
plt.bar(non_host['Country'],non_host['Bronze/Yr'],width=0.1,label='non-host')
plt.legend()
plt.show()
```

```python
#Medals won on an average as hosts and as non hosts comparison graphically:
plt.title("Medals won as host country and as non host country per year:")
X_axis = np.arange(len(host['Country']))
plt.bar(X_axis - 0.2,host['Total/Yr'],width=0.1,label='host-Total',color='yellow')
plt.bar(X_axis - 0.1,host['Gold/Yr'],width=0.1,label='host-Gold')
plt.bar(X_axis + 0.0,host['Silver/Yr'],width=0.1,label='host-Silver')
plt.bar(X_axis + 0.1,host['Bronze/Yr'],width=0.1,label='host-Bronze')
plt.xticks(X_axis, host['Country'])

plt.bar(X_axis - 0.2,non_host['Total/Yr'],width=0.1,label='non-host-Total')
plt.bar(X_axis - 0.1,non_host['Gold/Yr'],width=0.1,label='non-host-Gold')
plt.bar(X_axis + 0.0,non_host['Silver/Yr'],width=0.1,label='non-host-Silver')
plt.bar(X_axis + 0.1,non_host['Bronze/Yr'],width=0.1,label='non-host-Bronze')
plt.xticks(X_axis, non_host['Country'])

plt.legend()
plt.show()
```

**For Olymmpic Games, From the above graph, we can infer that the number of total medals and gold medals are highly affected by playing as host nation. Silver and Bronze are little affected but, they are still more than medals won as non host country.**

**Other asian countries have never been host in Olympic games from 1950 onwards. So their data is filtered out in this analysis by default.**
