
In this question the analysis is related to the players who won some medal in asian games but not able to get the same type of medal in olympic games and vice versa, so we are actualy trying to find the players let's say a player who won a gold medal in asian game but not able to won gold in olympic game.

by doing this kind of analysis we can find those players who win a gold medal in asian but not in olympic and vice versa, and we can call them in-consistent players who were not able to maintain consistency between asian and olympic games. Also we can do this analysis in two ways, first sportwise and second countrywise. By doing sportwise analysis we can find for a particular sport which players were not able to maintain the consistency between asian and olympic games. 

```python
import pandas as pd
import numpy as np
```

```python
oly_player=pd.read_csv("Players_Medal_List_Olympics.csv")
asian_player=pd.read_csv("Players_Medal_List_Asian.csv")
```

```python
oly_player.head()
```

```python
asian_player.head()
```

```python
oly_player2=oly_player.drop(columns=["Year","Host_City","Event","Sport_Type"])
```

```python
oly_player2.head()
```

```python
oly_player2=oly_player2.groupby(by=["Country","Sport","Gender","Name"]).sum().reset_index()
```

```python
oly_player2.head()
```

```python
asian_player2=asian_player.drop(columns=["Year","Event","Sport_Type"])
```

```python
asian_player2.head()
```

```python
asian_player2=asian_player2.groupby(by=["Country","Sport","Gender","Player_Name"]).sum().reset_index()
```

```python
asian_player2.head()
```

```python
asia_olympic=asian_player2.merge(oly_player2,left_on="Player_Name",right_on="Name",suffixes=('_Asian','_Olympic'))
```

```python
asia_olympic.head()
```

```python
sport_wise={}

for x in asian_player2["Sport"].unique():
    temp={}
    
    df1=asian_player2[asian_player2["Sport"]==x]
    df2=oly_player2[oly_player2["Sport"]==x]
    
    if len(df1)>0 and len(df2)>0:
        result=df1.merge(df2,left_on="Player_Name",right_on="Name",suffixes=('_Asian','_Olympic'))
        
        temp["Asian_Olympic_Gold"]=result[(result["Gold_Asian"]>=1) & (result["Gold_Olympic"]==0)][["Name","Country_Asian"]].values
        temp["Asian_Olympic_Silver"]=result[(result["Silver_Asian"]>=1) & (result["Silver_Olympic"]==0)][["Name","Country_Asian"]].values
        temp["Asian_Olympic_Bronze"]=result[(result["Bronze_Asian"]>=1) & (result["Bronze_Olympic"]==0)][["Name","Country_Asian"]].values
        
        temp["Olympic_Asian_Gold"]=result[(result["Gold_Asian"]==0) & (result["Gold_Olympic"]>=1)][["Name","Country_Asian"]].values
        temp["Olympic_Asian_Silver"]=result[(result["Silver_Asian"]==0) & (result["Silver_Olympic"]>=1)][["Name","Country_Asian"]].values
        temp["Olympic_Asian_Bronze"]=result[(result["Bronze_Asian"]==0) & (result["Bronze_Olympic"]>=1)][["Name","Country_Asian"]].values
        
        sport_wise[x]=temp
```

```python
sport_wise
```

as we can see that we get the list of players sport-wise,so for any particular sport we can find the players who were not able to maintain the consistency between asian and olympic games.Also if we want we can find for a particular country also the list of inconsistent players for a particular sport.

```python
max_players_gold=[0,'']
for x,y in sport_wise.items():
    cnt=0
    for z in y["Asian_Olympic_Gold"]:
        if z[1]=="India":
            cnt+=1
    for z in y["Olympic_Asian_Gold"]:
        if z[1]=="India":
            cnt+=1
    if cnt>max_players_gold[0]:
        max_players_gold[0]=cnt
        max_players_gold[1]=x
max_players_gold
```

So, as we can see that for india in the sport "Field hockey" there are maximum number of players who did not able to maintain the consistency between asian and olympic gold medals.

```python
players_list=[]
x=sport_wise[max_players_gold[1]]
for y in x["Asian_Olympic_Gold"]:
    if y[1]=="India":
        players_list.append([y[0],"Asian_Gold"])
for y in x["Olympic_Asian_Gold"]:
    if y[1]=="India":
        players_list.append([y[0],"Olympic_Gold"])
players_list
```

```python
asian_player3=asian_player2.drop(columns=["Sport"])
oly_player3=oly_player2.drop(columns=["Sport"])
```

```python
asian_player3=asian_player3.groupby(by=["Country","Player_Name"]).sum().reset_index()
oly_player3=oly_player3.groupby(by=["Country","Name"]).sum().reset_index()
```

```python
overall={}

result=oly_player3.merge(asian_player3,left_on="Name",right_on="Player_Name",suffixes=('_Olympic','_Asian'))

overall["Asian_Olympic_Gold"]=result[(result["Gold_Asian"]>=1) & (result["Gold_Olympic"]==0)][["Name","Country_Asian"]].values
overall["Asian_Olympic_Silver"]=result[(result["Silver_Asian"]>=1) & (result["Silver_Olympic"]==0)][["Name","Country_Asian"]].values
overall["Asian_Olympic_Bronze"]=result[(result["Bronze_Asian"]>=1) & (result["Bronze_Olympic"]==0)][["Name","Country_Asian"]].values

overall["Olympic_Asian_Gold"]=result[(result["Gold_Asian"]==0) & (result["Gold_Olympic"]>=1)][["Name","Country_Asian"]].values
overall["Olympic_Asian_Silver"]=result[(result["Silver_Asian"]==0) & (result["Silver_Olympic"]>=1)][["Name","Country_Asian"]].values
overall["Olympic_Asian_Bronze"]=result[(result["Bronze_Asian"]==0) & (result["Bronze_Olympic"]>=1)][["Name","Country_Asian"]].values
```

```python
overall
```

so here we can see that we get the overall list of all players,who were not able to maintain the consistency between asian and olympic games.Also if we want we can find for a particular country the list of inconsistent players.

```python
countries=result["Country_Olympic"].unique()
max_player_country=[]
for x in countries:
    arr1=overall["Asian_Olympic_Gold"]
    arr1=arr1[arr1[:,1]==x]
    
    arr2=overall["Olympic_Asian_Gold"]
    arr2=arr2[arr2[:,1]==x]
    max_player_country.append([x,len(arr1)+len(arr2)])

max_player_country=sorted(max_player_country,key=lambda x:x[1])
print(max_player_country[-1])
```

So as we can see that china is the country with maximum number of players,who were not able to maintain the consistency between asian and olympic games w.r.t. gold medal
