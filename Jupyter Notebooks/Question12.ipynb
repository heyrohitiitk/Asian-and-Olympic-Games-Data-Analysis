In this particular question we are trying to compare asian and olympic games by finding the players who won medals either gold or silver or bronze in both olympic and asian games.

Now by doing this analysis for any particluar country if we want to find those players who won gold in both asian and olymic then we can easily find and this will help us in identifying most valuable players.We can also find in any particular sport the players who won medal both in asian and olympic games, so this is a very useful insight which gives us the comparison of players between asian and olympic games and helps finding those players who perform well in both asian and olympic games.

```python
import pandas as pd
import numpy as np
```

```python
asian=pd.read_csv("Players_Medal_List_Asian.csv")
olympic=pd.read_csv("Players_Medal_List_Olympics.csv")
```

```python
asian2=asian[asian["Sport_Type"]=="Single"]
asian2=asian2.drop(columns=["Year","Event","Sport_Type"])
```

```python
asian2=asian2.groupby(by=["Country","Sport","Gender","Player_Name"]).sum().reset_index()
```

```python
asian2["Total"]=asian2["Gold"]+asian2["Silver"]+asian2["Bronze"]
```

```python
asian2.head()
```

```python
olympic2=olympic[olympic["Sport_Type"]=="Individual"]
olympic2=olympic2.drop(columns=["Year","Host_City","Event","Sport_Type"])
```

```python
olympic2=olympic2.groupby(by=["Country","Sport","Gender","Name"]).sum().reset_index()
```

```python
olympic2["Total"]=olympic2["Gold"]+olympic2["Silver"]+olympic2["Bronze"]
```

```python
olympic2.head()
```

```python
result=olympic2.merge(asian2,left_on="Name",right_on="Player_Name",suffixes=('_Olympic','_Asian'))
result.head()
```

```python
player_won_both=result[(result["Total_Olympic"]>=1) & (result["Total_Asian"]>=1)][["Name","Country_Olympic","Gender_Olympic","Sport_Olympic","Total_Asian","Total_Olympic"]]
player_won_both
```

so here we get the list of players who won a medal in both asian and olympic games, Now from this table we can derive many things like, we can get the list of players who won a medal in both asian and olympic games for a particular country and also we can find for a particular sport players who won a medal in both asian and olympic games, so this is a very informative insight,and we can call such countries or sports as effective in terms winning in both asian and olympic in which there are maximum numbers of players who won a medal in both asian and olympic games.

```python
player_won_both_india=player_won_both[player_won_both["Country_Olympic"]=="India"].drop(columns=["Country_Olympic"])
player_won_both_india
```

```python
player_won_gold=result[(result["Gold_Olympic"]>=1) & (result["Gold_Asian"]>=1)][["Name","Country_Olympic","Gender_Olympic","Sport_Olympic","Gold_Asian","Gold_Olympic"]]
player_won_gold
```

here we get those players who won gold medal in both asian and olympic games, and we can also do filtering here for a particular country or for a particular sport

```python
player_won_gold_india=player_won_gold[player_won_gold["Country_Olympic"]=="India"].drop(columns=["Country_Olympic"])
player_won_gold_india
```
