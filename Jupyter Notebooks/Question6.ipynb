```python
#Importing the required dataset:
df_o1=pd.read_csv('Players_Medal_list_Olympics.csv')
df_o2=pd.read_csv('Medal_List_Olympics.csv')
df_a1=pd.read_csv('Players_Medal_List_Asian.csv')
df_a2=pd.read_csv('Medal_List_Asian.csv')
```

# Q6: Sport wise top 5 athletes of various countries:

```python
#Same as above. Instead of inputs as [Country, Year]; now we'll have inputs as [Country, Sport, Year]
```

**Currently we'll work with top 5 athletes corresponding to each country, overall etc.**

```python
sp=input("Enter the sport:")
nat=input("Enter the Nation's name:")
yr=input("Enter the year:")
```

```python
if yr!='Overall':
    yr=int(yr)
```

```python
yr
```

```python
df_o1
```

```python
df1=df_o1[:]
```

```python
df1
```

```python
df1['Total']=df1['Gold']+df1['Silver']+df1['Bronze']
```

```python
df1
```

```python
if nat!='Overall':
    if yr!='Overall':
        df1=df1[df1['Sport']==sp]
        df1=df1[df1["Country"]==nat]
        df1=df1[df1['Year']==yr].reset_index(drop=True)
        df1=df1.groupby(['Name'])['Gold','Silver','Bronze','Total'].sum().reset_index()
        df1=df1.sort_values(['Gold','Silver','Bronze'], ascending=False)
        df1=df1.reset_index(drop=True)
        df1=df1[:5]
    else:
        df1=df1[df1['Sport']==sp]
        df1=df1[df1["Country"]==nat]
        df1=df1.groupby(['Name'])['Gold','Silver','Bronze','Total'].sum().reset_index()
        df1=df1.sort_values(['Gold','Silver','Bronze'], ascending=False)
        df1=df1.reset_index(drop=True)
        df1=df1[:5]
else:
    if yr!='Overall':
        df1=df1[df1['Sport']==sp]
        df1=df1[df1['Year']==yr]
        df1=df1.groupby(['Name','Gender','Country'])['Gold','Silver','Bronze','Total'].sum().reset_index()
        df1=df1.sort_values(['Gold','Silver','Bronze'], ascending=False)
        df1=df1.reset_index(drop=True)
        df1=df1[:5]   
    else:
        df1=df1[df1['Sport']==sp]
        df1=df1.groupby(['Name','Gender','Country'])['Gold','Silver','Bronze','Total'].sum().reset_index()
        df1=df1.sort_values(['Gold','Silver','Bronze'], ascending=False)
        df1=df1.reset_index(drop=True)
        df1=df1[:5]
```
