# Finding Recession period for countries

```python
def recession(x):
    rec_years= []
    temp = []
    last = float('inf')
    for i in x.index:
        medals = x.loc[i,'Total']
        year = x.loc[i,'Year']
        if medals<last:
            temp.append(year)
            last = medals
        else:
            if len(temp)>len(rec_years):
                rec_years = temp.copy()
            temp.clear()
            temp.append(year)
            last = medals
    if len(rec_years)>1:
        x['Year'] = str(rec_years[0])+" - "+str(rec_years[len(rec_years)-1])
    else:
        x['Year'] = int(x.loc[x[['Total']].idxmin(), 'Year'])
    x.drop(['Country','Total'],axis = 1, inplace =True)
    return x.iloc[0]
```

```python
# function to find the recession years of each country in asian or olympic games
def find_recession_years(df):
    df = df[['Year','Country','Total']]
    df = df.groupby(['Year','Country']).sum().reset_index()
    df_recession = df.groupby('Country').apply(recession).reset_index()
    return df_recession
```

# Finding Booming period for countries

```python
def boom(x):
    rec_years= []
    temp = []
    last = float('-inf')
    for i in x.index:
        medals = x.loc[i,'Total']
        year = x.loc[i,'Year']
        if medals>last:
            temp.append(year)
            last = medals
        else:
            if len(temp)>len(rec_years):
                rec_years = temp.copy()
            temp.clear()
            temp.append(year)
            last = medals
    if len(rec_years)>1:
        x['Year'] = str(rec_years[0])+" - "+str(rec_years[len(rec_years)-1])
    else:
        x['Year'] = int(x.loc[x[['Total']].idxmax(), 'Year'])
    x.drop(['Country','Total'],axis = 1, inplace =True)
    return x.iloc[0]
```

```python
# this function find booming period for countries in asian or olympic games
def find_boom_years(df):
    df = df[['Year','Country','Total']]
    df = df.groupby(['Year','Country']).sum().reset_index()
    df_boom = df.groupby('Country').apply(boom).reset_index()
    return df_boom
```
