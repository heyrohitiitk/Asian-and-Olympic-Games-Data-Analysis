```python
# getting urls of all sports from year 1954 to 1998
sports_links = []
for url in url_list:
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    sport_anchors = soup.findAll('a', href = True, text='details')
    links = []
    for anchor in sport_anchors:
        link = base_url+anchor['href']
        if link not in unstructured_links:
            links.append(link)
    sports_links.append(links)
```

# Finding Medal Tables for all sports of years 1954 to 2018

```python
not_found_links = []
for i in range(len(years)):
    for link in sports_links[i]:
        r = requests.get(link)
        soup = BeautifulSoup(r.text, 'html.parser')
        heading = soup.find('h1', attrs={'class':'firstHeading'}).text.split(" at ")[0]
        try:
            table = soup.find('table', attrs = {'class':'wikitable sortable plainrowheaders jquery-tablesorter'})
            df = pd.read_html(str(table))
            df[0]['Year'] = years[i]
            df[0]['Sport'] = heading
            df[0].to_csv("Medal_List_"+heading+"_"+str(years[i])+".csv",index = False)
            # print(heading)
        except:
            print("table not found for "+heading+ " in year "+str(years[i]))
            not_found_links.append(link)
f = open("medal_list_not_found_links.txt",'w')
for link in not_found_links:
    f.write(link+"\n")
f.close()
        
```

```python
for k in range(len(years)):
    for sport in sports_links[k]:
        r = requests.get(sport)
        soup = BeautifulSoup(r.text, 'html.parser')
        # getting sport name
        heading = soup.find('h1', attrs={'class':'firstHeading'}).text.split(" at ")[0]
        # scraping all tables with given class name
        tables = soup.findAll(attrs = {'class':'wikitable plainrowheaders'})
    
        for i in range(len(tables)):
            df = pd.DataFrame(columns=['Player_Name','Gender','Year','Country','Sport','Event','Sport_Type','Gold','Silver','Bronze'])
            # scraping all table rows
            tr_list = tables[i].findAll("tr")
            # removing header row from tr_list as we have custom header
            tr_list.pop(0)
            last_event = ""
            shared_medal_no = 3
            for tr in tr_list:
                # scraping all table definition in a particular row
                td_list = tr.findAll('td')
                if len(td_list)>4:
                    del td_list[2]
                    del td_list[3]
                    del td_list[4]

                if len(td_list)==4:
                    gender = "Unknown"
                    event = td_list[0].text.strip().replace("details","")
                    if 'Men' in event or ' men' in event:
                        gender = 'M'
                    if 'Women' in event or ' women' in event:
                        gender = 'F'
                    td_list.pop(0)
                    shared_medal_no = 3
                    medal_no = 1
                    for td in td_list:
                        # scraping all anchor tags in a particular cell
                        anchors = td.findAll('a')

                        if len(anchors)==0:
                            shared_medal_no = medal_no-1
                            print(shared_medal_no)
                            print("multiple winner in single event "+event+" ("+heading+") ("+str(i+1)+" table) of year "+str(years[k]))
                            medal_no+=1
                            continue

                        if(len(anchors))>2:
                            team = 'Team'
                        else:
                            team = 'Single'

                        # country name is just next to an image
                        try:
                            country = td.find('img').find_next('a').text
                        except:
                            print("Error in event "+event+" ("+heading+") ("+str(i+1)+" table) of year "+str(years[k]))
                            continue

                        players = []
                        p_links = []
                        # gettting all players names from anchors
                        for anchor in anchors:
                            if anchor.text!=country:
                                players.append(anchor.text)
                                p_links.append(base_url+anchor['href'])
                        # print(players)
                        for x in range(len(players)):
                            gender1 = gender
                            if (gender=='Unknown' and len(tables)>1 and i==2) or (gender=='Unknown' and len(tables)==1):
                                r1 = requests.get(p_links[x])
                                soup1 = BeautifulSoup(r1.text, 'html.parser')
                                para = soup1.find('p').text
                                if 'He ' in para or ' he ' in para or ' his ' in para:
                                    gender1 = 'M'
                                if 'She ' in para or ' she ' in para or ' her ' in para:
                                    gender1 = 'F'
                                if len(players)==1:
                                    gender = gender1

                            if medal_no == 1:
                                df.loc[len(df.index)]=[players[x], gender1, years[k], country, heading, event, team, 1, 0, 0]
                            if medal_no == 2:
                                df.loc[len(df.index)]=[players[x], gender1, years[k], country, heading, event, team, 0, 1, 0]
                            if medal_no == 3:
                                df.loc[len(df.index)]=[players[x], gender1, years[k], country, heading, event, team, 0, 0, 1]
                        medal_no +=1
                    

                if len(td_list)<3:
                    medal_no = shared_medal_no
                    for td in td_list:
                        # scraping all anchor tags in a particular cell
                        anchors = td.findAll('a')

                        if(len(anchors))>2:
                            team = 'Team'
                        else:
                            team = 'Single'

                        # country name is just next to an image
                        try:
                            country = td.find('img').find_next('a').text
                        except:
                            print("Error in event "+event+" ("+heading+") ("+str(i+1)+" table) of year "+str(years[k]))
                            continue
                        
                        players = []
                        p_links = []
                        # gettting all players names from anchors
                        for anchor in anchors:
                            if anchor.text!=country:
                                players.append(anchor.text)
                                p_links.append(base_url+anchor['href'])
                        # print(players)
                        for x in range(len(players)):
                            gender1 = gender
                            if (gender=='Unknown' and len(tables)>1 and i==2) or (gender=='Unknown' and len(tables)==1):
                                r1 = requests.get(p_links[x])
                                soup1 = BeautifulSoup(r1.text, 'html.parser')
                                para = soup1.find('p').text
                                if 'He ' in para or ' he ' in para or ' his ' in para:
                                    gender1 = 'M'
                                if 'She ' in para or ' she ' in para or ' her ' in para:
                                    gender1 = 'F'
                                if len(players)==1:
                                    gender = gender1
                                    
                            if medal_no == 1:
                                df.loc[len(df.index)]=[players[x], gender1, years[k], country, heading, event, team, 1, 0, 0]
                            if medal_no == 2:
                                df.loc[len(df.index)]=[players[x], gender1, years[k], country, heading, event, team, 0, 1, 0]
                            if medal_no == 3:
                                df.loc[len(df.index)]=[players[x], gender1, years[k], country, heading, event, team, 0, 0, 1]
                        medal_no +=2

            if len(tables)>1:
                if i==0:
                    df['Gender'] = 'M'
                    df.to_csv("Players_Medal_List_"+heading+"_Men_"+str(years[k])+".csv",index=False)
                if i==1:
                    df['Gender'] = 'F'
                    df.to_csv("Players_Medal_List_"+heading+"_Women_"+str(years[k])+".csv",index = False)
                if i==2:
                    df.to_csv("Players_Medal_List_"+heading+"_Mixed_"+str(years[k])+".csv",index = False)
            else:
                df.to_csv("Players_Medal_List_"+heading+"_"+str(years[k])+".csv",index=False)
```
